```python
import os
import pandas as pd
import struct
import matplotlib.pyplot as plt

root_dir=os.path.join('x:/Marphys_Archive','Data','OSNAP','RTEB1L-FetchAZA')
infile=os.path.join(root_dir,"Converted-20230907.csv")
outcsv=os.path.join(root_dir,"pressure-record-20230907.csv")
```

```python
## EXTRACT DIGIQUARTZ
outpath='DQZ.csv'
with open(infile, 'rt') as f, open(outpath, 'w') as outfile:
    for line in f:
        if 'DQZ' in line:
            outfile.write(line)
ds=pd.read_csv(outpath)
dqs=ds[["Record Time", "Pressure (kPa)","Serial Number"]]
dqs=dqs.rename(columns={"Pressure (kPa)": "Pressure Digiquartz (kPa)"})
```

```python
## EXTRACT RS485 Presens
outpath='PRESENS.csv'
with open(infile, 'rt') as f, open(outpath, 'w') as outfile:
    for line in f:
        if 'PNS' in line:
            outfile.write(line)
ds=pd.read_csv(outpath)
pns=ds[["Record Time", "Pressure (kPa)","Serial Number"]]
pns=pns.rename(columns={"Pressure (kPa)": "Pressure Presens (kPa)"})
```

```python
infile
```

```python
## MERGE DATA AND PRINT TO CSV
fetch=pns.merge(dqs, how='inner',on='Record Time')
fetch.to_csv(outcsv)
```

```python
ds=fetch.set_index('Record Time')
ds=ds[["Pressure Digiquartz (kPa)","Pressure Presens (kPa)"]]
```

```python
ds["Pressure Digiquartz (kPa)"]=ds["Pressure Digiquartz (kPa)"]*0.1
ds["Pressure Presens (kPa)"]=ds["Pressure Presens (kPa)"]*0.1
ds.loc['17/07/2022 08:00':'05/09/2023 08:00'].plot(
                                                   figsize=(16,4),
                                                   grid=True,legend=True,
                                                  xlabel="",
                                                  ylabel="Pressure (decibar)",
                                                   linewidth=0.75,
                                                  style=['-k','-b'],alpha=0.5)
plt.tight_layout()
plt.savefig(os.path.join(root_dir,'RTEB1-pressure-first-pass.png'), dpi=300, facecolor='white')
```
