```python
import os
import pandas as pd
import struct
import matplotlib.pyplot as plt
root_dir=os.path.join('c:/Users','sa01ld','OneDrive - SAMS','Instruments','FETCH_AZA_M','Data','RTEBL2','June24-TC2014')
infile=os.path.join(root_dir,"Data_240626104123__007C14_2312_.csv")
outcsv=os.path.join(root_dir,"pressure-record-240626104123__007C14_2312.csv")
```

```python
## EXTRACT DIGIQUARTZ
outpath='DQZ.csv'
with open(infile, 'rt') as f, open(outpath, 'w') as outfile:
    for line in f:
        if 'DQZ' in line:
            outfile.write(line)
ds=pd.read_csv(outpath)
dqs=ds[["Record Time", "Pressure (kPa)","Serial Number"]]
dqs=dqs.rename(columns={"Pressure (kPa)": "Pressure Digiquartz (kPa)"})
```

```python
## EXTRACT RS485 Presens
outpath='PRESENS.csv'
with open(infile, 'rt') as f, open(outpath, 'w') as outfile:
    for line in f:
        if 'PNS' in line:
            outfile.write(line)
ds=pd.read_csv(outpath)
pns=ds[["Record Time", "Pressure (kPa)","Serial Number"]]
pns=pns.rename(columns={"Pressure (kPa)": "Pressure Presens (kPa)"})
```

```python
## EXTRACT BATTERY
outpath='BAT.csv'
with open(infile, 'rt') as f, open(outpath, 'w') as outfile:
    for line in f:
        if 'BAT' in line:
            outfile.write(line)
ds=pd.read_csv(outpath)
ds=ds.set_index('Record Time')

ds.loc['2022/07/17 08:00':'2023/09/05 08:00'].plot(y="Volts V",
                                                   figsize=(16,4),
                                                   grid=True,legend=True,
                                                  xlabel="",
                                                  ylabel="Volts",
                                                   linestyle = 'None',
                                                   marker="*",
                                                 alpha=0.5)
plt.savefig(os.path.join(root_dir,'Battery.png'), dpi=300, facecolor='white')
```

```python
## EXTRACT INCLINOMETER
outpath='INC.csv'
with open(infile, 'rt') as f, open(outpath, 'w') as outfile:
    for line in f:
        if 'INC' in line:
            outfile.write(line)
ds=pd.read_csv(outpath)
ds=ds.set_index('Record Time')
```

```python
ds.loc['2022/07/17 08:00':'2023/09/05 08:00'].plot(y=["Pitch (deg)","Roll (deg)"],
                                                   figsize=(16,4),
                                                   grid=True,legend=True,
                                                   xlabel="",
                                                   ylabel="Degrees",
                                                   linestyle = 'None',
                                                   marker=".",
                                                   alpha=0.5)
plt.title("Inclinometer")
plt.savefig(os.path.join(root_dir,'Inclinometer.png'), dpi=300, facecolor='white')
```

```python
## MERGE DATA AND PRINT TO CSV
fetch=pns.merge(dqs, how='inner',on='Record Time')
fetch.to_csv(outcsv)
fetch
```

```python
ds=fetch.set_index('Record Time')
ds=ds[["Pressure Digiquartz (kPa)","Pressure Presens (kPa)"]]
ds
```

```python
ds["Pressure Digiquartz (kPa)"]=ds["Pressure Digiquartz (kPa)"]*0.1
ds["Pressure Presens (kPa)"]=ds["Pressure Presens (kPa)"]*0.1
ds.loc['2024/04/20 08:00':'2024/06/25 08:00'].plot(
                                                   figsize=(16,4),
                                                   grid=True,legend=True,
                                                  xlabel="",
                                                  ylabel="Pressure (decibar)",
                                                   linewidth=0.75,
                                                  style=['-k','-b'],alpha=0.5)
plt.tight_layout()
plt.savefig(os.path.join(root_dir,'RTEB1-pressure-first-pass.png'), dpi=300, facecolor='white')
```

```python
da=ds["Pressure Presens (kPa)"]-ds["Pressure Digiquartz (kPa)"]
da.loc['2024/04/20 08:00':'2024/06/25 08:00'].plot(figsize=(16,4),
                                                   grid=True,legend=False,
                                                  xlabel="",
                                                  ylabel="Pressure difference (decibar)",
                                                  linewidth=0.5,color='b',alpha=0.8)
plt.title('Presens-Digiquartz')
plt.savefig(os.path.join(root_dir,'Presens-Digiquartz-diff.png'), dpi=300, facecolor='white')
```
