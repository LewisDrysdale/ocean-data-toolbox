```python
import numpy as np
from keras.preprocessing.text import Tokenizer

# Sample text data
text_data = ["This is some text data", "Here is another piece of text data", "And one more text data point"]

# Create a tokenizer object
tokenizer = Tokenizer()

# Fit the tokenizer on the text data
tokenizer.fit_on_texts(text_data)

# Convert text data to numerical arrays
numerical_data = tokenizer.texts_to_matrix(text_data)

print(numerical_data)
```
